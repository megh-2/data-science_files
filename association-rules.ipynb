```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
titanic= pd.read_csv(r"C:\Users\91863\Downloads\Titanic.csv")
titanic.head()
```

```python
df = pd.get_dummies(titanic,dtype =int)
```

```python
df
```

# apriori algo

```python
frequent_itemsets = apriori(df,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rule=association_rules(frequent_itemsets,metric = "lift",min_threshold=0.7)
rule
```

a leverage value of 0 indiactes independence. range will be [-1,1]
a high conviction value means that the consequent is highly depending on the antecedants and confidence

```python
rule.sort_values("lift",ascending=False)[0:20]
```

```python
rule[rule.lift>1]
```
