```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.manifold import  TSNE
from bioinfokit.visuz import cluster
```

```python
df=pd.read_csv(r"C:\Users\91863\Downloads\TSNE_data.csv")
df.head()
```

```python
# split-out valda dataset

array = df.values

# seperate array into input and output components

x= array[:,2:]
y= array[:,0]
```

```python
# TSNE visualization

from bioinfokit.visuz import cluster
data_tsne = TSNE(n_components =2).fit_transform(x)
cluster.tsneplot(score=data_tsne)
```

```python
# get a list of categories
color_class = df["diagnosis"].to_numpy()
cluster.tsneplot(score= data_tsne,colorlist = color_class,legendpos =  "upper right")
```
