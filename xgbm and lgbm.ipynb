```python
import numpy as np  
import pandas as pd
import matplotlib.pyplot as plt 
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv(r"C:\Users\91863\Downloads\pima-indians-diabetes.data.csv")
```

```python
data
```

```python
data.columns = ["preq","plas","pres","skin","test","mass","pedi","age","class"]
```

```python
data
```

```python
x = data.iloc[:,0:8]
y = data.iloc[:,8]
```

```python
# split data into train and test set
seed = 7
test_size =0.33
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=test_size,random_state = seed,shuffle = True)
```

```python
# fit the model no training data
model = XGBClassifier(max_depth=5)
model.fit(x_train,y_train)
```

```python
# make prediction fro test data
y_pred = model.predict(x_test)
predictions  = [round(value) for value in y_pred]
```

```python
from sklearn.metrics import accuracy_score
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test,predictions)
print("accuracy : ", (accuracy * 100.0))
```

# lgbm

```python
# spliiting  the data
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =00.25,random_state=0,shuffle =True)
```

```python
import lightgbm as lgb 
d_train = lgb.Dataset(x_train,label = y_train)
```

```python
params = {}
params["learning_rate"] = 0.003
params["boosting_type"] = "gbdt"
params["objective"] = "binary"
params["metric"] = "binary+logloss"
params["sub_feature"] = 0.5
params["num_leaves"] = 10
params["min_data"] = 50
params["max_depth"]=10
```

```python
clf = lgb.train(params,d_train,1500)
```

```python
# prediction
y_pred = clf.predict(x_test)
predictions= [round(value) for value in y_pred]
accuracy = accuracy_score(y_test,predictions)
```
