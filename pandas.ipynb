```python
traingle1
```

```python
def tritype(row):
    if (row["side1"] == row["side2"] == row["side3"]):
        return "equilateral"
    elif (row["side1"] == row["side2"]) or (row["side1"] == row["side3"]) or (row["side2"] == row["side3"]):
        return "isosceles" 
    else:
        return "scalene"  
```

```python
traingle["type"]=traingle.apply(tritype,axis=1)
```
