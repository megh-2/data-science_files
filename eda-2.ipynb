# isolation Forest

```python
from sklearn.ensemble import IsolationForest
```

```python
# training model 
clf = IsolationForest(random_state=10,contamination=0.01)
clf.fit(data_encoded) # ranodom state - consistency ,contamination = % of data as outliers
```

```python
# predictions

y_pred_outliers = clf.predict(data_encoded)
```

```python
# -1 for outliers and 1 for inliers
y_pred_outliers
```

```python
data_encoded
```

```python
# lets add a new data point which is outlier

data_encoded.loc[150] = [151,20,30,40,50,1,0,0]
```

```python
data_encoded
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=0.01)
clf.fit(data_encoded)
#predictions

y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
```

```python
data_encoded["scores"] = clf.decision_function(data_encoded)
```

```python
data_encoded["anomaly"] = clf.decision_function(data_encoded.iloc[:,0:8])
```
