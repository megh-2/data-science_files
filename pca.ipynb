```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import scale
from sklearn.decomposition import PCA
```

```python
univ = pd.read_csv(r"C:\Users\91863\Downloads\Universities (1).csv")
univ.head()
```

```python
univ.describe()
```

```python
# considering only numerical data  for scaling

univ.data=univ.iloc[:,1:]
univ.data.head()
```

```python
# convering into numpy array
UNI = univ.data.values
UNI
```

```python
uni_normal = scale(UNI)
uni_normal
```

```python
pca =PCA()
pca_values = pca.fit_transform(uni_normal)
```

```python
pca_values
```

```python
pca = PCA(n_components = 6)
pca_values  = pca.fit_transform(uni_normal)
```

```python
# the amount of varinaces that each PCA explains is

var= pca.explained_variance_ratio_
var
```

```python
# cumulative variance
var1 = np.cumsum(np.round(var,decimals=4)*100)
var1
```

```python
pca.components_
```

```python
# variance plot for PCA components obtained
plt.plot(var1,color="red")
```

```python
pca_values[:,0:1]
```

```python
# plot between PCA1 and PCA2

x = pca_values[:,0:1]
y = pca_values[:,1:2]
#z = pca_values[:,2:3]
plt.scatter(x,y)
```

```python
pca_values  = pd.DataFrame(pca_values[:,0:2],columns = ["pca1","pca2"])
```

```python
final_df = pd.concat([univ[["Univ"]],pca_values],axis=1)
```
